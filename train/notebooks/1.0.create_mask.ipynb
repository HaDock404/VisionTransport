```python
from PIL import Image, ImageDraw
```

```python
img = Image.new("RGB", (2048, 1024), (250, 170, 30, 255))
img.save(f'./test_mask/full_yellow.png')
img
```

```python
img = Image.new("RGB", (2048, 1024), (0, 0, 142, 255))
img.save(f'./test_mask/full_blue.png')
img
```

```python
img = Image.new("RGB", (2048, 1024), "black")
draw = ImageDraw.Draw(img)
polygons = [(0, 0), (0, 1024), (1024, 1024), (1024, 0)]
draw.polygon(polygons, fill=(250, 170, 30, 255))
polygons = [(1025, 0), (2048, 0), (2048, 1024), (1025, 1024)]
draw.polygon(polygons, fill=(0, 0, 142, 255))
img.save(f'./test_mask/yellow_blue.png')
img
```
