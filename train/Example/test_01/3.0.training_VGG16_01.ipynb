```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from skimage.transform import resize
from tensorflow.keras.layers import Reshape
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
```

```python
X_train = df_train['Image_Path']
y_train = df_train['Target_Path']

X_test = df_val['Image_Path']
y_test = df_val['Target_Path']
```

```python
def preprocess_image(image_path):
    img = load_img(image_path, target_size=(256, 256))
    img_array = img_to_array(img)
    return img_array

X_train_processed = np.array([preprocess_image(img_path) for img_path in X_train])
X_test_processed = np.array([preprocess_image(img_path) for img_path in X_test])
```

```python
def preprocess_label(image_path):
    img = load_img(image_path, target_size=(256, 256))  
    img_array = img_to_array(img)
    return img_array

y_train_processed = np.array([preprocess_label(img_path) for img_path in y_train])
y_test_processed = np.array([preprocess_label(img_path) for img_path in y_test])
```

```python
num_classes_train = len(np.unique(y_train_processed))
num_classes_test = len(np.unique(y_test_processed))

print("Nombre de classes dans les données d'entraînement :", num_classes_train)
print("Nombre de classes dans les données de test :", num_classes_test)
```

```python
import numpy as np

# 1. Identifier les classes uniques dans vos données
unique_classes_train = np.unique(y_train_processed)
unique_classes_test = np.unique(y_test_processed)

print("Classes uniques dans les données d'entraînement :", unique_classes_train)
print("Classes uniques dans les données de test :", unique_classes_test)

# 2. Regrouper certaines classes en nouvelles classes
# Par exemple, vous pouvez diviser vos classes en 10 groupes
# Voici un exemple de regroupement, mais vous pouvez ajuster cela selon vos besoins

def group_classes(classes, num_groups):
    grouped_classes = np.array_split(classes, num_groups)
    return grouped_classes

num_groups = 10
grouped_classes_train = group_classes(unique_classes_train, num_groups)
grouped_classes_test = group_classes(unique_classes_test, num_groups)

# 3. Mettre à jour les étiquettes de vos données pour refléter ces nouvelles classes
def update_labels(classes, labels):
    updated_labels = np.zeros_like(labels)
    for i, group in enumerate(classes):
        for original_label in group:
            updated_labels[np.where(labels == original_label)] = i
    return updated_labels

y_train_updated = update_labels(grouped_classes_train, y_train_processed)
y_test_updated = update_labels(grouped_classes_test, y_test_processed)

# 4. Visualiser à nouveau le nombre de classes uniques
num_updated_classes_train = len(np.unique(y_train_updated))
num_updated_classes_test = len(np.unique(y_test_updated))

print("Nombre de classes uniques après regroupement dans les données d'entraînement :", num_updated_classes_train)
print("Nombre de classes uniques après regroupement dans les données de test :", num_updated_classes_test)
```

```python
y_train_encoded = to_categorical(y_train_updated, num_classes=10)
y_test_encoded = to_categorical(y_test_updated, num_classes=10)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, UpSampling2D, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint

# Définir l'architecture du modèle U-Net
def unet_model(input_shape=(256, 256, 3), num_classes=10):
    inputs = Input(input_shape)

    # Encodeur
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    conv2 = Conv2D(128, 3, activation='relu', padding='same')(pool1)
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    # Chemin central
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(pool2)
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(conv3)

    # Décodeur
    up4 = concatenate([UpSampling2D(size=(2, 2))(conv3), conv2], axis=-1)
    conv4 = Conv2D(128, 3, activation='relu', padding='same')(up4)
    conv4 = Conv2D(128, 3, activation='relu', padding='same')(conv4)

    up5 = concatenate([UpSampling2D(size=(2, 2))(conv4), conv1], axis=-1)
    conv5 = Conv2D(64, 3, activation='relu', padding='same')(up5)
    conv5 = Conv2D(64, 3, activation='relu', padding='same')(conv5)

    # Couche de sortie
    outputs = Conv2D(num_classes, 10, activation='softmax')(conv5)

    model = Model(inputs=inputs, outputs=outputs)
    return model

# Créer le modèle U-Net
model = unet_model()

# Compiler le modèle
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Créer un générateur de données pour l'entraînement
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Créer un générateur de données pour la validation
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

# Définir les générateurs de données pour l'entraînement et la validation
train_generator = train_datagen.flow(X_train_processed, y_train_encoded, batch_size=32)
validation_generator = test_datagen.flow(X_test_processed, y_test_encoded, batch_size=32)

# Définir les callbacks
checkpoint = ModelCheckpoint("unet_segmentation.keras", monitor='val_accuracy', verbose=1, save_best_only=True, save_weights_only=False, mode='auto')

y_train_adjusted = np.expand_dims(y_train_encoded, axis=-2)
y_test_adjusted = np.expand_dims(y_test_encoded, axis=-2)

# Entraîner le modèle avec les étiquettes ajustées
history = model.fit(train_generator,
                    steps_per_epoch=len(X_train_processed) // 32,
                    epochs=20,
                    validation_data=validation_generator,
                    validation_steps=len(X_test_processed) // 32,
                    callbacks=[checkpoint])

```
