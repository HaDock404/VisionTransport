```python
from tensorflow.keras.models import load_model

model = load_model('./models/model_01.h5')
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

# Charger l'image de test que vous souhaitez segmenter
new_image = image.load_img(
    "./Data/files/Cityscapes_leftImg8bit_trainvaltest/leftImg8bit/test/berlin/berlin_000000_000019_leftImg8bit.png",
    target_size=(224, 224))

# Convertir l'image en tableau numpy
new_image_array = image.img_to_array(new_image)

# Ajouter une dimension supplémentaire pour le batch
new_image_batch = np.expand_dims(new_image_array, axis=0)

# Faire une prédiction sur l'image avec votre modèle
prediction = model.predict(new_image_batch)

# Récupérer l'index de la classe prédite pour chaque pixel
predicted_class = np.argmax(prediction[0], axis=-1)

# Définir une palette de couleurs pour chaque classe
colors = {
    0: [0, 0, 0],       # Classe 0: Fond (Noir)
    1: [255, 0, 0],     # Classe 1: Bâtiments (Rouge)
    2: [0, 255, 0],     # Classe 2: Routes (Vert)
    # Ajoutez plus de classes avec leurs couleurs correspondantes si nécessaire
}

# Créer un masque coloré en fonction de la prédiction
colored_mask = np.zeros_like(new_image_array)
for c in range(len(colors)):
    colored_mask[predicted_class == c] = colors[c]

# Afficher l'image d'origine avec le masque superposé
plt.imshow(new_image)
plt.imshow(colored_mask, alpha=0.5)
plt.title('Segmentation Predicted')
plt.axis('off')
plt.show()
```
